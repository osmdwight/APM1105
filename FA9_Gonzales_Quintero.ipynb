```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_spotify = pd.read_csv('C:/Users/DWIGHT JIRO/Downloads/Spotify_Youtube.csv')
df_spotify.head(10)
```

```python
#3
print(df_spotify.columns)
```

```python
#4

#no. of columns
rows= len(df_spotify.axes[0])
print("Number of rows:",rows)

#data types
print("Column names and data types:")
print(df_spotify.dtypes)

#no. of data points
columns = len(df_spotify.axes[1])
num_data_points = rows * columns

print("\nNumber of data points:", num_data_points)

#null
print("\nColumns with null values:")
print(df_spotify.isnull().sum())
```

```python
df_spotify[['Likes', 'Comments']] = df_spotify[['Likes', 'Comments']].replace(0, np.nan)


print(df_spotify.head())
```

```python
#6

df_spotify.describe()
```

```python
#7

album_count = (df_spotify['Album_type'] == 'album').sum()
single_count = (df_spotify['Album_type'] == 'single').sum()
compilation_count = (df_spotify['Album_type'] == 'compilation').sum()

print("Total number of albums: ", album_count)
print("Total number of singles: ", single_count)
print("Total number of compilations: ", compilation_count)
```

```python
#8


pie_data = df_spotify['Album_type'].value_counts()


plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%')


plt.title('Album types')


plt.show()





```

```python
#9

grouped_data = df_spotify.groupby('Artist')[['Views', 'Stream']].sum()
print(grouped_data)

```

```python
top_youtube = df_spotify.groupby('Artist')['Views'].sum().nlargest(10)
top_spotify = df_spotify.groupby('Artist')['Stream'].sum().nlargest(10)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))


ax1.bar(top_youtube.index, top_youtube.values)
ax1.set_title('Top 10 Artists on Youtube')
ax1.set_xlabel('Artist')
ax1.set_ylabel('Total Views')


ax2.bar(top_spotify.index, top_spotify.values)
ax2.set_title('Top 10 Artists on Spotify')
ax2.set_xlabel('Artist')
ax2.set_ylabel('Total Streams')


plt.tight_layout()
plt.show()
```

```python
#11
artist_rankings = df_spotify.groupby('Artist')[['Views', 'Stream']].sum()
artist_sorted = artist_rankings.sort_values(['Views', 'Stream'], ascending=False)
top_10 = artist_sorted.head(10)
top_10

```

```python
#12
top_10_views = df_spotify.nlargest(10, 'Views')
print(top_10_views)
# Top 10 songs based on comments
top_10_comments = df_spotify.nlargest(10, 'Comments')
print(top_10_comments)
# Top 10 songs based on likes
top_10_likes = df_spotify.nlargest(10, 'Likes')
print(top_10_likes)
```

```python
#13

top_10_views.plot(kind='barh', x='Track', y='Views', title='Top 10 songs with having the most views')
top_10_comments.plot(kind='barh', x='Track', y='Comments', title='Top 10 songs with having the most comments')
top_10_likes.plot(kind='barh', x='Track', y='Likes', title='Top 10 songs with having the most likes')
```

```python
#14
top_10_songs = df_spotify.sort_values('Stream', ascending=False).head(10)
top_10_songs[['Track', 'Energy', 'Danceability', 'Acousticness']]
```

```python

#16

artist_likes = df_spotify.groupby('Artist')['Likes'].sum()
top_10_artists = artist_likes.nlargest(10)
top_10_artists.plot(kind='barh')
plt.xlabel('Number of Likes')
plt.ylabel('Artist')
plt.title('Top 10 Artists by Total Number of Likes')
plt.show()
```

```python
#17


singles = df_spotify[df_spotify["Album_type"] == "single"]
grouped = singles.groupby("Artist").sum()[["Views", "Likes"]]
top_10_artists = grouped.nlargest(10, "Likes")
print(top_10_artists)
```

```python
#20



df_channels = df_spotify.groupby("Channel")["Views"].sum()
df_sort_channels = df_channels.nlargest(10)
sns.set_style("whitegrid")
sns.barplot(x=df_sort_channels.values, y=df_sort_channels.index, color='c')
sns.despine(left=True)
plt.title('Top Spotify Channels based on their Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```
